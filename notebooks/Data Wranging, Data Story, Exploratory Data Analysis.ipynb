# Cleaning up dataframe to filter out rows where year = 0

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from pandas.plotting import lag_plot
from decimal import Decimal
%matplotlib inline
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv('10_metadata.csv', names=['Track_ID','Title','Song_ID','Release','Artist_ID','Artist_MBID','Artist_Name','duration','Artist_Familiarity','Artist_Hotttnesss','year','Track_7digitalid','shs_perf','shs_work'])
df.index += 1
df.head()
```

```python
df = df[['duration', 'year']]
df.head()
```

```python
df['year'] = df['year'].replace({'0':np.nan, 0:np.nan})
df
```

```python
df = df[pd.notnull(df['year'])]
df
```

```python
df[df.duration != 'duration']
```
